Hieronder krijg je de Python functie `gradient_descent` cadeau dat het Gradient Descent algoritme implementeert. Deze functie hebben we ook in de les gezien, alleen benadert deze functie de afgeleide numeriek m.b.v. functie `df` i.p.v. dat je zelf de afgeleide moet opgeven. De afgeleide wordt als volgt benaderd:

> $\frac{df}{dx} \approx \frac{f(x+\epsilon) - f(x)}{\epsilon}$


```python
def gradient_descent(f, x0, alpha=0.1, epsilon=1e-5, delta=1e-9, maxiter=1000):
    '''
    zoekt het minimum van functie f m.b.v. gradient descent
    f is een functie
    x0 is de startwaarde
    alpha is de learning rate (default 0.1)
    epsilon is het eindige verschil om de afgeleide te berekenen (default 1e-5)
    delta is het stopcriterium (default 1e-9)
    maxiter is het maximum aantal iteraties (default 1000)
    '''
    x = [x0]  # alle x-waarden bijhouden in een list
    i = 1
    while np.abs(df(f, x0, epsilon)) > delta and i <= maxiter:
        x0 -= alpha * df(f, x0, epsilon)
        x.append(x0)
        i += 1
    print('aantal iteraties:', i)
    print('minimum gevonden in x =', x[-1])
    return x

def df(f, x, epsilon):
    '''
    benadert de afgeleide van functie f in het punt x d.m.v. een eindig verschil
    f is een functie
    x is de functiewaarde
    epsilon is het eindige verschil om de afgeleide te berekenen: (f(x+epsilon) - f(x)) / epsilon
    '''
    y, y_eps = f(x), f(x + epsilon)
    return (y_eps - y) / epsilon
```
